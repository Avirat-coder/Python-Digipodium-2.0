```python
import pandas as ps 
import seaborn as sns
```

```python
df = sns.load_dataset("titanic")
df
```

```python
fig = sns.histplot(df,x="age",kde=True,color="blue",edgecolor="black")
fig.set_title("age distibution of titanic passenger  ")
fig.set_xlabel("age")
fig.set_ylabel("fequency")
```

```python
fig= sns.countplot(df, x = 'survived',palette = 'RdBu', hue= 'sex')
fig.set_title("Survival count of Titantic Passenger")
fig.set_xlabel("survived")
fig.set_ylabel("count")
```

```python
sns.barplot(df,x='sex',y='fare',hue='class')
fig.set_title("Average Fare by Passenger Class and Gender ")
fig.set_xlabel("sex")
fig.set_ylabel("Average Fare")
```

```python
fig=sns.countplot(df,x='class',palette='viridis')
fig.set_title("count of passenger by class")
fig.set_xlabel("passenger class")
fig.set_ylabel("count")
```

```python
sns.pairplot(df[['age','fare','survived','pclass']],hue='survived',palette='terrain')
```
